*   برخی از کشورها مانند ایران و مصر تعداد کل مرگ و میر بیشتری را در مقایسه با سایر کشورهای منطقه نشان می دهند.

* همچنین ممکن است عواملی مانند زیرساخت های مراقبت های بهداشتی، مداخلات دولت، و شناخت وبررسی بیماری  به تفاوت های مشاهده شده بین کشورها کمک کنند.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a professional style with Seaborn
sns.set(style="whitegrid")

# Read the CSV file
data = pd.read_csv('full_grouped.csv')

# General plot function for reusability
def plot_bar_chart(data, title, xlabel, ylabel, color_palette, rotation=45):
    plt.figure(figsize=(12, 6))
    bars = plt.bar(data.index, data.values, color=sns.color_palette(color_palette, len(data)))
    
    # Add value labels on top of bars
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', fontsize=9)
    
    plt.title(title, fontsize=14, weight='bold')
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    plt.xticks(rotation=rotation, ha='right')
    plt.tight_layout()
    plt.show()

# 1. Total Deaths in Eastern Mediterranean Region by Country
mid_co = data[data['WHO Region'] == 'Eastern Mediterranean']
deaths_per_country = mid_co.groupby('Country/Region')['Deaths'].sum()

plot_bar_chart(
    deaths_per_country, 
    title='Total Deaths in Eastern Mediterranean Region by Country', 
    xlabel='Country', 
    ylabel='Total Deaths', 
    color_palette='Blues'
)
```

* داده‌ها می‌توانند هم شدت شیوع و هم تفاوت در استراتژی‌های واکنش را در این کشورها منعکس کنند.



```python
# 2. Total Number of Deaths by Country (Europe)
europe_countries = ['United Kingdom', 'Italy', 'France', 'Spain', 'Russia', 'Belgium', 'Netherlands', 'Sweden', 'Turkey']
europe_data = data[data['Country/Region'].isin(europe_countries)]
deaths_per_country_europe = europe_data.groupby('Country/Region')['Deaths'].sum()

plot_bar_chart(
    deaths_per_country_europe, 
    title='Total Number of Deaths by Country (Europe)', 
    xlabel='Country', 
    ylabel='Total Deaths', 
    color_palette='Reds'
)
```

**نتیجه گیری برای ابتلا کووید-۱۹ در برزیل، مکزیک، ایالات متحده، بریتانیا**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.dates import DateFormatter

# Set a professional style with Seaborn
sns.set(style="whitegrid")

# Load the dataset
file_path = 'total_cases.csv'
data = pd.read_csv(file_path)

# Convert 'date' column to datetime format for easier handling
data['date'] = pd.to_datetime(data['date'])

# Filter the data for the required countries and take one day from each month for plotting
monthly_data = data.set_index('date').resample('M').last()

# Create a figure with 4 subplots (3rd subplot)
fig, axs = plt.subplots(1, 4, figsize=(20, 6))  # 1 row, 4 columns for each country

# List of countries for the third subplot
countries_line = ['Brazil', 'Mexico', 'United States', 'United Kingdom']

# Custom line styles and colors
line_styles = ['-', '--', '-.', ':']
markers = ['o', 's', '^', 'd']
colors = sns.color_palette('Set2', len(countries_line))

for i, country in enumerate(countries_line):
    axs[i].plot(
        monthly_data.index, 
        monthly_data[country], 
        label=country, 
        linestyle=line_styles[i], 
        marker=markers[i], 
        color=colors[i]
    )
    
    # Set titles and labels
    axs[i].set_title(f'COVID-19 Infected in {country}', fontsize=12, weight='bold')
    axs[i].set_xlabel('Date', fontsize=10)
    axs[i].set_ylabel('Deaths', fontsize=10)
    
    # Add gridlines and legend
    axs[i].grid(True, which='both', linestyle='--', linewidth=0.5)
    axs[i].legend(loc='upper left', fontsize=9)

    # Format x-axis to show dates clearly
    axs[i].xaxis.set_major_formatter(DateFormatter('%b %Y'))
    axs[i].tick_params(axis='x', rotation=45)

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
```

**نتیجه‌ گیری برای ابتلا به کووید-۱۹ در ایران، عربستان سعودی، ژاپن و چین**

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns

# Set a professional Seaborn style
sns.set(style="whitegrid")

# Load the dataset
file_path = 'total_cases.csv'  # Your dataset path
data = pd.read_csv(file_path)

# Convert 'date' column to datetime format
data['date'] = pd.to_datetime(data['date'])

# Filter the data to only include the 5th day of each month
data['day'] = data['date'].dt.day
data_5th = data[data['day'] == 5]  # Only rows where the day is the 5th
data_5th = data_5th.set_index('date')

# Interpolate to fill any missing data for Saudi Arabia
data_5th['Saudi Arabia'] = data_5th['Saudi Arabia'].interpolate()

# Create a figure with 4 subplots (1st subplot)
fig, axs = plt.subplots(1, 4, figsize=(20, 6))  # 1 row, 4 columns for each country

# Specify the countries for the bar plot
countries_bar = ['Iran', 'Saudi Arabia', 'Japan', 'China']

# Color palette
colors = sns.color_palette('Set2', len(countries_bar))

# Plot bar charts for each country with enhanced aesthetics
for i, country in enumerate(countries_bar):
    axs[i].bar(data_5th.index, data_5th[country], color=colors[i], width=20, edgecolor='black', alpha=0.8)
    
    # Set title and labels with custom font sizes
    axs[i].set_title(f'COVID-19 Deaths in {country}', fontsize=14, weight='bold')
    axs[i].set_xlabel('Date', fontsize=12)
    axs[i].set_ylabel('Deaths (Log Scale)', fontsize=12)

    # Set y-axis to be logarithmic
    axs[i].set_yscale('log')

    # Set x-axis to show only the 5th day of every 3rd month for clearer spacing
    axs[i].xaxis.set_major_locator(mdates.MonthLocator(interval=3))  # Show every 3rd month
    axs[i].xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))  # Format as 'Month Year'

    # Rotate the date labels for better visibility
    axs[i].tick_params(axis='x', rotation=45)

    # Add gridlines for better readability
    axs[i].grid(True, which='both', linestyle='--', linewidth=0.7)

# Adjust layout to avoid overlap and make it more visually appealing
plt.tight_layout()

# Show the plots
plt.show()
```

نمودارها  مقایسه واضحی از شدت و مدیریت بیماری همه گیر کووید-19 در این چهار کشور ارائه می دهد. آمار تلفات در ایران به طور قابل توجهی بیشتر از سایر کشورها است که نشان دهنده چالش های مداوم در کنترل این همه گیری است. عربستان سعودی و ژاپن موفق شدند نرخ مرگ و میر خود را پایین نگه دارند، در حالی که افزایش اولیه چین با اقدامات کنترلی موثری دنبال شد که از تلفات بیشتر در مقیاس بزرگ جلوگیری کرد. تفاوت در آمار مرگ و میر این کشورها، تفاوت‌ها را در پاسخ‌های بهداشت عمومی، ظرفیت مراقبت‌های بهداشتی و سیاست‌های دولت در سراسر همه‌گیری نشان می‌دهد.

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt


tickers = ['^GSPC', '^DJI', 'CL=F', 'GC=F', 'DX-Y.NYB']


data = yf.download(tickers, start='2020-01-01', end='2021-12-31')['Adj Close']
data.head()
```

تجزیه و تحلیل عملکرد بازار سهام

*  تجزیه تحلیل چگونگی واکنش شاخص های سهام در طول کووید 19
* S&P 500 and Dow Jones

*   دو شاخصی که نوسانات قابل توجهی را در طول دوره همه گیر نشان می دهند. S&P 500 و Dow Jones
*   بازار سهام در ابتدا در اوایل سال 2020 به دلیل قرنطینه ها افت شدیدی را تجربه کرد، اما به ویژه در بخش فناوری و مراقبت های بهداشتی به شدت بازگشت و به افزایش شاخص ها کمک کرد.

**اقتصاد چین**

داده های سهام، نفت خام، طلا و نرخ ارز برای چین از سال 2019 تا 2022

*   عملکرد بازار سهام (شاخص ترکیبی SSE): بازار همراه با فراز و نشیب اما انعطاف پذیر و امیدوار کننده.
*   کالاها (نفت خام و طلا): نفت کاهش شدید قیمت را تجربه کرد، در حالی که طلا به عنوان یک دارایی امن با ثبات باقی ماند.
* نرخ مبادله ارز (USD/CNY): یوان چین در برابر دلار آمریکا نوساناتی را تجربه کرد، اما در مقایسه با سایر شاخص ها نسبتاً ثابت ماند.


```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np


symbols = {
    'SSE Composite Index': '000001.SS',
    'Crude Oil': 'CL=F',
    'Gold': 'GC=F',
    'USD/CNY': 'CNY=X'
}

data = {}
for name, symbol in symbols.items():
    data[name] = yf.download(symbol, start='2019-01-01', end='2022-12-31')


def plot_with_stats(df, title, ylabel):
    mean_val = df['Close'].mean()
    std_val = df['Close'].std()

    plt.figure(figsize=(10, 6))
    plt.plot(df.index, df['Close'], label=title)
    plt.axhline(mean_val, color='r', linestyle='--', label=f'Mean: {mean_val:.2f}')
    plt.axhline(mean_val + std_val, color='g', linestyle='--', label=f'Mean + STD: {mean_val + std_val:.2f}')
    plt.axhline(mean_val - std_val, color='b', linestyle='--', label=f'Mean - STD: {mean_val - std_val:.2f}')

    plt.fill_between(df.index, mean_val - std_val, mean_val + std_val, color='gray', alpha=0.2)
    plt.title(title)
    plt.ylabel(ylabel)
    plt.xlabel('Date')
    plt.legend()
    plt.grid(True)
    plt.show()


plot_with_stats(data['SSE Composite Index'], 'Shanghai Stock Exchange (SSE) Index', 'Index Value')

plot_with_stats(data['Crude Oil'], 'Crude Oil Prices (WTI)', 'Price (USD)')

plot_with_stats(data['Gold'], 'Gold Prices (USD)', 'Price (USD)')

plot_with_stats(data['USD/CNY'], 'USD/CNY Exchange Rate', 'Exchange Rate (CNY/USD)')
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np


start_date = '2020-01-01'
end_date = '2021-01-01'


symbols = {
    'DAX': '^GDAXI',
    'Oil': 'CL=F',
    'Gold': 'GC=F',
    'EUR/USD': 'EURUSD=X'
}


data_germany = {symbol: yf.download(ticker, start=start_date, end=end_date)['Adj Close'] for symbol, ticker in symbols.items()}


fig, axs = plt.subplots(2, 2, figsize=(15, 10))

axs[0, 0].plot(data_germany['DAX'], label='DAX', color='blue')
mean_dax = data_germany['DAX'].mean()
std_dax = data_germany['DAX'].std()
axs[0, 0].axhline(mean_dax, color='green', linestyle='--', label='Mean')
axs[0, 0].axhline(mean_dax + std_dax, color='red', linestyle='--', label='Mean + Std')
axs[0, 0].axhline(mean_dax - std_dax, color='red', linestyle='--', label='Mean - Std')
axs[0, 0].set_title('DAX Index (Germany)')
axs[0, 0].set_xlabel('Date')
axs[0, 0].set_ylabel('Price')
axs[0, 0].legend()


axs[0, 1].plot(data_germany['Oil'], label='Crude Oil (WTI)', color='orange')
mean_oil = data_germany['Oil'].mean()
std_oil = data_germany['Oil'].std()
axs[0, 1].axhline(mean_oil, color='green', linestyle='--', label='Mean')
axs[0, 1].axhline(mean_oil + std_oil, color='red', linestyle='--', label='Mean + Std')
axs[0, 1].axhline(mean_oil - std_oil, color='red', linestyle='--', label='Mean - Std')
axs[0, 1].set_title('Crude Oil Prices (WTI)')
axs[0, 1].set_xlabel('Date')
axs[0, 1].set_ylabel('Price')
axs[0, 1].legend()


axs[1, 0].plot(data_germany['Gold'], label='Gold', color='gold')
mean_gold = data_germany['Gold'].mean()
std_gold = data_germany['Gold'].std()
axs[1, 0].axhline(mean_gold, color='green', linestyle='--', label='Mean')
axs[1, 0].axhline(mean_gold + std_gold, color='red', linestyle='--', label='Mean + Std')
axs[1, 0].axhline(mean_gold - std_gold, color='red', linestyle='--', label='Mean - Std')
axs[1, 0].set_title('Gold Prices')
axs[1, 0].set_xlabel('Date')
axs[1, 0].set_ylabel('Price')
axs[1, 0].legend()


axs[1, 1].plot(data_germany['EUR/USD'], label='EUR/USD', color='purple')
mean_eur_usd = data_germany['EUR/USD'].mean()
std_eur_usd = data_germany['EUR/USD'].std()
axs[1, 1].axhline(mean_eur_usd, color='green', linestyle='--', label='Mean')
axs[1, 1].axhline(mean_eur_usd + std_eur_usd, color='red', linestyle='--', label='Mean + Std')
axs[1, 1].axhline(mean_eur_usd - std_eur_usd, color='red', linestyle='--', label='Mean - Std')
axs[1, 1].set_title('EUR/USD Exchange Rate')
axs[1, 1].set_xlabel('Date')
axs[1, 1].set_ylabel('Rate')
axs[1, 1].legend()
plt.tight_layout()
plt.show()
```
